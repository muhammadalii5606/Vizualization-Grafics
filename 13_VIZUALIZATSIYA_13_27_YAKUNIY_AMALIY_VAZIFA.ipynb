### **Ustunlar tavsifi**  

1. **Country** – mamlakat nomi  
   - Bu satrda mamlakatlarning nomlari joylashgan.  

2. **Happiness.Rank** – baxtlilik darajasi bo‘yicha reyting  
   - Har bir mamlakatning baxt darajasi bo‘yicha egallagan o‘rni.  
   - **1** – eng baxtli mamlakat, **so‘nggi o‘rin** – eng baxtsiz mamlakat.  

3. **Happiness.Score** – baxtlilik darajasi  
   - Mamlakatning umumiy baxtlilik darajasi **(0–10 oralig‘ida)**.  
   - Bu baho turli omillarga asoslangan holda berilgan.  

4. **Whisker.high** – yuqori ishonch oralig‘i (Upper Confidence Interval)  
   - **Happiness Score** qiymatining yuqori chegarasi (ishonchlilik darajasi bilan).  

5. **Whisker.low** – pastki ishonch oralig‘i (Lower Confidence Interval)  
   - **Happiness Score** qiymatining pastki chegarasi.  

6. **Economy..GDP.per.Capita.** – iqtisodiy farovonlik (YAIM)  
   - Aholi jon boshiga to‘g‘ri keladigan yalpi ichki mahsulot (YAIM).  
   - Mamlakat iqtisodining baxtlilik darajasiga ta’sirini ko‘rsatadi.  

7. **Family** – jamiyatning qo‘llab-quvvatlashi  
   - Ijtimoiy aloqalar va oilaviy qo‘llab-quvvatlash darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar o‘zini shunchalik baxtli his qiladi.  

8. **Health..Life.Expectancy.** – sog‘lom umr davomiyligi  
   - Odamlarning o‘rtacha qancha umr ko‘rishi va sog‘lom hayot kechirishi darajasi.  
   - Tibbiy xizmat sifati va hayot sharoitlari bilan bog‘liq.  

9. **Freedom** – hayot tanlovlarida erkinlik  
   - Odamlarning o‘z hayotini erkin boshqarish darajasi.  
   - Mamlakatdagi demokratik va fuqarolik erkinliklarini aks ettiradi.  

10. **Generosity** – saxiylik darajasi  
   - Jamiyatdagi saxiylik va xayriya qilishga moyillik darajasi.  
   - Ushbu qiymat qanchalik yuqori bo‘lsa, odamlar ko‘proq yordam berishga tayyor bo‘ladi.  

11. **Trust..Government.Corruption.** – korrupsiya haqidagi tasavvurlar  
   - Mamlakatdagi korrupsiya darajasi bo‘yicha odamlarning fikri.  
   - **0 ga yaqin bo‘lsa**, odamlar davlatga ishonmaydi.  
   - **Yuqori qiymat bo‘lsa**, davlatga ishonch baland bo‘ladi.  

12. **Dystopia.Residual** – baxtiyorlik darajasiga ta’sir etmaydigan omillar  
   - Ideal bo‘lmagan (Dystopia) sharoitlarda baxtlilik darajasini oshirishga yordam beruvchi faktorlar.  
   - Bu ko‘rsatkich mamlakat baxtlilik darajasi **modelining qoldiq qiymati** sifatida tushuniladi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/happines_rank_2017.csv"
df = pd.read_csv(url)
df.head()
```

#1) Baxtlilik darajasi va aholi jon boshiga to‘g‘ri keladigan YAIM o‘rtasidagi bog‘liqlikni scatterplot bilan vizualizatsiya qiling.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Economy..GDP.per.Capita.', y='Happiness.Score')

plt.title('Baxtlilik darajasi va YAIM o\'rtasidagi bog\'liqlik')
plt.xlabel('Aholi jon boshiga to‘g‘ri keladigan YAIM')
plt.ylabel('Baxtlilik darajasi')
plt.grid(True)
plt.show()
```

#2) Baxtlilik darajasi (`Happiness.Score`) taqsimotini histogram orqali chizing.

```python
plt.figure(figsize=(10, 6))

sns.histplot(df['Happiness.Score'], bins=20, kde=True, color='skyblue')

plt.title('Baxtlilik darajasi (Happiness Score) taqsimoti', fontsize=14)
plt.xlabel('Baxtlilik balli', fontsize=12)
plt.ylabel('Davlatlar soni', fontsize=12)

plt.grid(axis='y', alpha=0.3)
plt.show()
```

#3) Eng baxtli va eng baxtsiz 10 ta mamlakatni barplot orqali ko‘rsating.

```python
top_10 = df.nlargest(10, 'Happiness.Score')
bottom_10 = df.nsmallest(10, 'Happiness.Score')

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(data=top_10, x='Happiness.Score', y='Country', ax=ax[0], palette='viridis')
ax[0].set_title('Top 10: Eng baxtli mamlakatlar', fontsize=14)
ax[0].set_xlim(0, 8)

sns.barplot(data=bottom_10, x='Happiness.Score', y='Country', ax=ax[1], palette='magma')
ax[1].set_title('Bottom 10: Eng baxtsiz mamlakatlar', fontsize=14)
ax[1].set_xlim(0, 8)

plt.tight_layout()
plt.show()
```

#4)  Baxtlilik darajasi (`Happiness.Score`) va ijtimoiy omillar (Freedom, Family Support, Generosity) bilan scatterplot (`subplot` orqali ketma ket) chizing!

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

sns.scatterplot(ax=axes[0], data=df, x='Freedom', y='Happiness.Score', color='teal', alpha=0.7)
axes[0].set_title('Erkinlik va Baxtlilik', fontsize=14)
axes[0].set_xlabel('Erkinlik darajasi')
axes[0].set_ylabel('Baxtlilik balli')

sns.scatterplot(ax=axes[1], data=df, x='Family', y='Happiness.Score', color='coral', alpha=0.7)
axes[1].set_title('Oila (Ijtimoiy ko\'mak) va Baxtlilik', fontsize=14)
axes[1].set_xlabel('Oila/Ijtimoiy qo\'llab-quvvatlash')
axes[1].set_ylabel('')

sns.scatterplot(ax=axes[2], data=df, x='Generosity', y='Happiness.Score', color='green', alpha=0.7)
axes[2].set_title('Saxovat va Baxtlilik', fontsize=14)
axes[2].set_xlabel('Saxovatlilik darajasi')
axes[2].set_ylabel('')

plt.tight_layout()
plt.show()
```

#5) Korrupsiya va baxtlilik darajasi bog‘liqligini koʻrsating!

```python
plt.figure(figsize=(10, 6))

sns.regplot(data=df, x='Trust..Government.Corruption.', y='Happiness.Score',
            scatter_kws={'alpha':0.6, 'color':'darkblue'},
            line_kws={'color':'red'})

plt.title('Korrupsiya (Hukumatga ishonch) va Baxtlilik o\'rtasidagi bog\'liqlik', fontsize=14)
plt.xlabel('Hukumatdagi korrupsiyaning pastligi (Ishonch darajasi)', fontsize=12)
plt.ylabel('Baxtlilik balli', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

#6) Baxtlilik darajasi va YAIM o‘rtasidagi bog‘liqlikni chiziqli regressiya orqali tekshiring!

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['Economy..GDP.per.Capita.']].values
y = df['Happiness.Score'].values

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Haqiqiy ma\'lumotlar')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regressiya chizig\'i')

r_sq = model.score(X, y)
plt.text(0.1, 7.5, f'R^2 koeffitsiyenti: {r_sq:.3f}', fontsize=12, color='darkgreen', weight='bold')

plt.title('YAIM va Baxtlilik: Chiziqli Regressiya modeli', fontsize=14)
plt.xlabel('Aholi jon boshiga YAIM')
plt.ylabel('Baxtlilik balli')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print(f"Model kesishish nuqtasi (b0): {model.intercept_}")
print(f"Model koeffitsiyenti (b1): {model.coef_[0]}")
```

#7) Issiqlik xaritasi (heatmap) orqali korrelyatsiyalarni ko‘rsating.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Omillar o\'rtasidagi korrelyatsiya (Heatmap)', fontsize=15)
plt.show()
```

#8) 10 ta eng yuqori va eng past YAIMga ega mamlakatlarni barplot orqali aks ettiring!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

top_gdp = df.nlargest(10, 'Economy..GDP.per.Capita.')

bottom_gdp = df.nsmallest(10, 'Economy..GDP.per.Capita.')

fig, ax = plt.subplots(2, 1, figsize=(12, 12))

sns.barplot(data=top_gdp, x='Economy..GDP.per.Capita.', y='Country', ax=ax[0], palette='Blues_r')
ax[0].set_title('Top 10: Eng yuqori YAIMga ega mamlakatlar', fontsize=14)
ax[0].set_xlabel('YAIM ko\'rsatkichi')

sns.barplot(data=bottom_gdp, x='Economy..GDP.per.Capita.', y='Country', ax=ax[1], palette='Reds_r')
ax[1].set_title('Bottom 10: Eng past YAIMga ega mamlakatlar', fontsize=14)
ax[1].set_xlabel('YAIM ko\'rsatkichi')

plt.tight_layout()
plt.show()
```

#9) Oilaviy qo‘llab-quvvatlash va YAIM o‘rtasidagi bog‘liqlikni aks ettiring!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.jointplot(data=df, x='Economy..GDP.per.Capita.', y='Family',
                  kind='reg', height=8, ratio=5, color="purple",
                  scatter_kws={'alpha':0.5}, line_kws={'color':'orange'})

g.fig.suptitle('YAIM va Oilaviy qo\'llab-quvvatlash o\'rtasidagi bog\'liqlik', y=1.02, fontsize=14)
g.set_axis_labels('Aholi jon boshiga YAIM', 'Oila / Ijtimoiy qo\'llab-quvvatlash', fontsize=12)

plt.show()
```


# 10)  Saxiylik (Generosity) va hukumat korrupsiyasi (Trust..Government.Corruption.) o‘rtasidagi bog‘liqlikni scatterplot yordamida vizualizatsiya qiling!

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Trust..Government.Corruption.', y='Generosity',
                hue='Happiness.Score', size='Happiness.Score',
                palette='coolwarm', sizes=(20, 200), alpha=0.7)

plt.title('Saxiylik va Hukumatga ishonch o\'rtasidagi bog\'liqlik', fontsize=14)
plt.xlabel('Hukumatdagi korrupsiyaning pastligi (Ishonch)', fontsize=12)
plt.ylabel('Saxiylik darajasi', fontsize=12)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title='Baxt balli')

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#11) Baxtlilik darajasi bo‘yicha mamlakatlarni “Juda baxtli”, “Baxtli”, “O‘rtacha”, “Baxtsiz” toifalariga ajrating va har bir guruh uchun o‘rtacha iqtisodiy ko‘rsatkichlarni (GDP per Capita) ko‘rsating.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = [0, 4, 5.5, 6.5, 10]
labels = ['Baxtsiz', 'O‘rtacha', 'Baxtli', 'Juda baxtli']

df['Toifa'] = pd.cut(df['Happiness.Score'], bins=bins, labels=labels)

gdp_by_category = df.groupby('Toifa')['Economy..GDP.per.Capita.'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=gdp_by_category, x='Toifa', y='Economy..GDP.per.Capita.', palette='viridis')

plt.title('Baxtlilik toifalari bo‘yicha o‘rtacha YAIM ko‘rsatkichi', fontsize=14)
plt.xlabel('Baxtlilik toifasi', fontsize=12)
plt.ylabel('O‘rtacha YAIM (GDP per Capita)', fontsize=12)

for i, val in enumerate(gdp_by_category['Economy..GDP.per.Capita.']):
    plt.text(i, val + 0.02, f'{val:.2f}', ha='center', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

#12) Muhim o‘zgaruvchilar (GDP, Life Expectancy, Freedom, Generosity, Happiness Score) bo‘yicha pairplot orqali umumiy taqsimotni vizualizatsiya qiling.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_plot = [
    'Happiness.Score',
    'Economy..GDP.per.Capita.',
    'Health..Life.Expectancy.',
    'Freedom',
    'Generosity'
]

g = sns.pairplot(df[columns_to_plot],
                 kind='reg',
                 diag_kind='kde',
                 plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.4}},
                 palette='husl')

g.fig.suptitle('Muhim o‘zgaruvchilar bo‘yicha umumiy bog‘liqlik va taqsimot (Pairplot)', y=1.02, fontsize=16)

plt.show()
```
